### Handling NaN using KNN Imputation

```python
from sklearn.impute import KNNImputer
```

```python
knn = KNNImputer(n_neighbors=3,add_indicator=True)
knn.fit(df)
df = knn.transform(df)
```

```python
df
```

## Modelling - Hierarchical Clustering

#### Plotting Dendrogram for no of Clusters

```python
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(10,7))
plt.title("Years Dendograms")
dend = shc.dendrogram(shc.linkage(df, method='ward'))
```

#### From Dendrogram --> Clusters=2

### Model Building

```python
from sklearn.cluster import AgglomerativeClustering

cluster= AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
labels = cluster.fit_predict(df)
```

```python
labels
```

```python
plt.figure(figsize=(10,7))
plt.scatter(df[:,1],df[:,2], c=cluster.labels_, cmap='rainbow')
```
